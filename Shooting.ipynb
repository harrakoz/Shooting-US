```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

from collections import Counter
```

```python
plt.figure(figsize = (16, 16))
plt.style.use("bmh")
csv_file = pd.read_csv("shootings.csv")
```

```python
grouped = csv_file.groupby("race")
```

```python
plt.hist(csv_file["race"], 20, facecolor="orange")
plt.xlabel("Race")
plt.ylabel("Frequency")
plt.title("Frequency of The most shooted races")

plt.savefig("Frequency of the most shooted races in a histogram.png")
```

```python
csv_file
```

```python
counted = Counter(np.array(csv_file["race"]))
summed_values = sum(counted.values())
frequency_race = {}
```

```python
for key, value in counted.items():
    frequency_race[key] = value / summed_values * 100
frequency_race
plt.pie(counted.values(), labels=counted.keys(), autopct=lambda x: str(round(x, 2))+"%")
plt.title("Race Frequencies")
plt.legend(loc="upper left", fontsize="x-small")
```

```python
race_data = []
for race in list(csv_file["race"]):
    race_data.append(frequency_race[race])
csv_file["frequency_race"] = race_data
```
